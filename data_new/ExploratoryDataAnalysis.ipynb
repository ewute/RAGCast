```python
import pandas as pd
# Libraries for plots
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Read CSV into a Pandas dataframe
df = pd.read_csv("epic_all_songs_lines_allspeakers.csv")
```

```python
sns.countplot(x = "speaker", data = df)
plt.xticks(rotation=90)
plt.xlabel("Speaker")
plt.ylabel("Count")
plt.tight_layout()
plt.title("Fig. 1: Lines Spoken per Character")
plt.show()
```

```python
counts = df['speaker'].value_counts().reset_index()
counts.columns = ['speaker', 'count']
fig = px.bar(
    counts.sort_values('count', ascending=True),  # ascending=True  least at bottom, most at top
    x = 'count',
    y= 'speaker',
    title='Fig. 1: Lines Spoken per Character',
    orientation = 'h'
)
fig.update_layout(
    xaxis_title='Count',
    yaxis_title='Speaker',
    height=800,  # increase if you have many speakers
    margin=dict(l=150, r=50, t=80, b=50)
)


fig.show()
```

```python
df['word_count'] = df['line'].str.split().str.len()
avg_words = (df.groupby('speaker')['word_count'].mean().reset_index().sort_values('word_count', ascending=True))
#print(avg_words)

fig = px.bar(
    avg_words, 
    x = 'word_count',
    y = 'speaker',
    title = "Fig.2: Average Words per Line by Character",
)

fig.update_layout(
    xaxis_title = 'Average Words per Line',
    yaxis_title = 'Character',
    height=800,  # increase if you have many speakers
    margin=dict(l=150, r=50, t=80, b=50)
)
```
